```python
import numpy as np
import matplotlib.pyplot as plt
from statistics import mean

np.random.seed(7)
#xs = np.random.randint(1, 10, 10)
#ys = np.random.randint(30, 50, 10)

xs = np.array([1, 2, 3, 4, 5])
ys = np.array([3, 5, 6, 7, 7])

print(f"x={xs}")
print(f"y={ys}")

def best_of_fit(xs, ys):
        m = ((mean(xs) * mean(ys)) - mean(xs*ys)) / (mean(xs)**2 - mean(xs**2))
        b = mean(ys) - m * mean(xs)
        return m, b
m, b = best_of_fit(xs, ys)
print(m, b)

regression_line = [(m * x) + b for x in xs]

future_x = np.array([6, 7, 8, 9, 10])
future_y = m * future_x + b

plt.scatter(x=xs, y=ys)
plt.scatter(future_x, future_y)
plt.plot(xs, regression_line)
plt.show()
```

```python
#r_squared
def squared_error(ys_orig, ys_line):
    return sum((ys_line-ys_orig)**2)
def coef_det(ys_orig, ys_line):
    ys_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, ys_mean_line)
    return 1 - (squared_error_regr / squared_error_y_mean)

r_squared = coef_det(ys, regression_line)
print(r_squared)
```
